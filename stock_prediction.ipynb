```python
df.loc[:, ['Adj Close', 'Volume']].head()
```

```python
df_reg = df.loc[:, ['Adj Close', 'Volume']]
df_reg['HL_PCT'] = (df['High'] - df['Low']) / df['Close'] * 100
df_reg['PCT_change'] = (df['Close'] - df['Open']) / df['Open'] * 100
df_reg.head()
```

```python
forecast_out = int(math.ceil(0.01 * len(df_reg)))
print(forecast_out)
forecast_col = 'Adj Close'
df_reg.head()
```

```python
df_reg.fillna(value=-9999, inplace=True)
print(df_reg.shape)
df_reg['label'] = df_reg[forecast_col].shift(-forecast_out)
X = np.array(df_reg.drop(['label'], 1))
X = preprocessing.scale(X)
X_lately = X[-forecast_out:]
X = X[:-forecast_out]

y = np.array(df_reg['label'])
y = y[:-forecast_out]
print('Dimension of X', X.shape)
print('Dimension of y', y.shape)
```

```python
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Linear Regression usage
l_reg = LinearRegression(n_jobs=-1)
l_reg.fit(X_train, y_train)
l_confidence_reg = l_reg.score(X_test, y_test)
forecast_set = l_reg.predict(X_lately)
df_reg['forecast'] = np.nan
print(forecast_set, forecast_out)
```

```python
last_date = df_reg.iloc[-1].name
last_unix = last_date
next_unix = last_unix + datetime.timedelta(days=1)

for i in forecast_set:
    next_date = next_unix
    next_unix += datetime.timedelta(days=1)
    df_reg.loc[next_date] = [np.nan for _ in range(len(df_reg.columns)-1)]+[i]

df_reg['Adj Close'].tail(500).plot()
df_reg['forecast'].tail(500).plot()
plt.legend(loc=4)
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Quadatric Regression 2 usage
quad2_reg = make_pipeline(PolynomialFeatures(2), Ridge())
quad2_reg.fit(X_train, y_train)
quad2_conf_reg = quad2_reg.score(X_test, y_test)
```
